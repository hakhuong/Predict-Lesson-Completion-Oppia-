```python
import numpy as np
import pandas as pd
import seaborn as sns


from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('./Data/datastudiodf.csv')
```

```python
data.isna().sum()
```

# Logistic regression

- Dependent variable: Exploration_completed (binary variable)
- Independent variables: 

```python
regr_df = data.copy()
```

```python
regr_df = regr_df[['medium', 'device_cat', 'new_users', 'exploration_completed', 'avg_session_duration', 
                   'country_segment', 'pagedepth_1', 
                   'landing_page_type_2']]
```

```python
regr_df.sample(5)
```

```python
# Convert boolean into 1 0 
regr_df['new_users'] = regr_df['new_users'].astype('int64')
regr_df['exploration_completed'] = regr_df['exploration_completed'].astype('int64')
regr_df['pagedepth_1'] = regr_df['pagedepth_1'].astype('int64')
```

```python
# Convert categorical columns to dummy variable
regr_df = pd.get_dummies(regr_df)
```

```python
regr_df.sample(5)
```

```python
X = regr_df.drop(['exploration_completed'], axis=1)
y = regr_df['exploration_completed']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=123)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = pd.Series(model.predict(X_test))
y_test = y_test.reset_index(drop=True)
z = pd.concat([y_test, y_pred], axis=1)
z.columns = ['True', 'Prediction']
z.head()
```

```python
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
print("Precision:", metrics.precision_score(y_test, y_pred))
print("Recall:", metrics.recall_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```
